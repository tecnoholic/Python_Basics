```python
# VISIONADO DE UN CONJUNTO DE COMPASES
# ------------------------------------
from music21 import * 
obra = converter.parse('e:/Tomas/Basura/Bach Bourree I - BWV1009.mxl', format='musicxml')

# Vamos a ver sólo los primeros 8 compases (todos los compases de la obra menos el último):
CompasesDeLaObra = obra.measures(1, 8)
CompasesDeLaObra.show()
```

```python
# CONTAR NOTAS Y COMPASES
# -----------------------
from music21 import * 
obra = converter.parse('e:/Tomas/Basura/Bach Bourree I - BWV1009.mxl', format='musicxml')

# Para analizar una obra, es interesante hacer que todos los datos sean 'plano', es decir,
# que la información de las notas no esté incluida dentro de compases o dentro de partes.

ObraPlana = obra.flat
print ("La obra tiene", len(ObraPlana.notes), "notas.")

for thisNote in ObraPlana:
    print(thisNote.step)
```

```python
# LAS NOTAS DE LA OBRA (TONO Y DURACIÓN)
# --------------------------------------
from music21 import * 
obra = converter.parse('e:/Tomas/Basura/Bach Bourree I - BWV1009.mxl', format='musicxml')

for thisNote in obra.notes:
    print(thisNote.octave)
```

```python
# LA TONALIDAD DE LA OBRA
# -----------------------
from music21 import * 
obra = converter.parse('e:/Tomas/Basura/Bach Bourree I - BWV1009.mxl', format='musicxml')

print ("La tonalidad es", obra.analyze('key'), end='.\n')

# Los tonos (sonidos) mínimo y máximo de la obra se calculan así:
p = analysis.discrete.Ambitus()
print ("En la obra se encuentran sonidos cuya frecuencia está en este rango:", p.getPitchRanges(obra))
print ("En la obra se encuentran notas comprendidas en este rango:", p.getPitchSpan(obra))

# Es curioso que aunque la obra está en Re mayor (D major), la armadura es la de Sol mayor.
```
